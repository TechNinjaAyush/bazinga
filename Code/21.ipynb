```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load dataset
df = pd.read_csv("../files/adult_dataset.csv")
df.head()
```

```python
# Set styles
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
#Histogram - Distribution of Age
plt.figure()
sns.histplot(df['age'], bins=30, kde=True)
plt.title("Histogram: Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Dot Plot - Frequency of Education Levels
plt.figure()
edu_counts = df['education'].value_counts().sort_values()
plt.plot(edu_counts.values, edu_counts.index, 'o')
plt.title("Dot Plot: Frequency of Education Levels")
plt.xlabel("Count")
plt.ylabel("Education Level")
plt.grid(True)
plt.show()
```

```python
# Bar Plot - Average Hours-per-Week by Income Group
plt.figure()
sns.barplot(data=df, x='income', y='hours-per-week', estimator=np.mean)
plt.title("Bar Plot: Average Hours Worked per Week by Income Group")
plt.xlabel("Income Group")
plt.ylabel("Average Hours per Week")
plt.show()
```

```python
# i. Line Chart - Trend of Average Hours-per-Week vs Age
plt.figure()
age_hours = df.groupby('age')['hours-per-week'].mean()
plt.plot(age_hours.index, age_hours.values, color='green')
plt.title("Line Chart: Average Hours-per-Week by Age")
plt.xlabel("Age")
plt.ylabel("Average Hours per Week")
plt.grid(True)
plt.show()
```

```python
# j. Box Plot with Histogram and Scatter
g = sns.JointGrid(data=df, x="age", y="hours-per-week", height=8)
g.plot_joint(sns.scatterplot, alpha=0.4)
g.plot_marginals(sns.histplot, kde=True)
g.ax_joint.set_title("Box+Histogram+Scatter: Age vs Hours-per-Week", pad=80)
plt.show()
```
