```python
import pandas as pd
```

```python
# load dataset
movies=pd.read_csv('../files/movies.csv')
movies.head()
```

```python
# Create data subsets for different languages (Original Language)

english_movies=movies[movies['original_language']=='en']
korean_movies=movies[movies['original_language']=='ko']
```

```python
english_movies.head()
```

```python
korean_movies.head()
```

```python
# Merge two subsets
merged_movies=pd.concat([english_movies,korean_movies])
merged_movies.head()
```

```python
# Sort Data using customer ratings.

movies.columns.tolist()
```

```python
sorted_movies=movies.sort_values(by='popularity',ascending=False)
sorted_movies.head()
```

```python
# Transposing Data

transposed_movies=movies.head().T
transposed_movies.head()
```

```python
# Melting Data to Long Format

melted_movies=pd.melt(movies,id_vars=['title','original_language'],var_name='attribute',value_name='value')
melted_movies.head()
```

```python
# Casting Data to Wide Format

casted_movies = melted_movies.pivot_table(index='original_language', 
                                          columns='attribute', 
                                          values='value', 
                                          aggfunc=lambda x: x.iloc[0] if x.dtype == 'O' else x.mean())
casted_movies.head()
```
