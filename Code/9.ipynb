```python
import pandas as pd
```

```python
#load the dataset

df=pd.read_csv("../files/dataset_Facebook.csv")
df.head(10)
```

```python
link_posts=df[df['Type']=='Link']
video_posts=df[df['Type']=='Video']
photo_posts=df[df['Type']=='Photo']
status_posts=df[df['Type']=='Status']
```

```python
print("Link Posts:\n")
link_posts.head()
```

```python
print("Video Posts:\n")
video_posts.head()
```

```python
merged_subset=pd.concat([link_posts,video_posts],ignore_index=True)
print("Merged Subset")
merged_subset.head(10)
```

```python
sorted_df=df.sort_values(by='Page total likes',ascending=False)
print("Sorted Data")
sorted_df.head(10)
```

```python
transposed_df=df.transpose()
print("Transposed Dataframe")
transposed_df.head(10)
```

```python
melted_df=pd.melt(df, id_vars=['Type'], value_vars=['like', 'share', 'comment'],var_name='Metric', value_name='Value')
print("Melted DataFrame")
melted_df.head()
```

```python
wide_df=melted_df.pivot_table(index=['Type'],columns='Metric',values='Value',aggfunc='mean').reset_index()
print("Wide format DataFrame:")
wide_df.head()
```
