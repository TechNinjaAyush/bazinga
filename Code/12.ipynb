```python
import pandas as pd
```

```python
# load dataset

adult=pd.read_csv("../files/adults.csv")
adult.head()
```

```python
# Create data subsets for different Country, Sex, race.

india_subset = adult[adult['native-country'] == 'India']
female_subset = adult[adult['sex'] == 'Female']
white_race_subset = adult[adult['race'] == 'White']

print("\nIndia Subset:\n", india_subset.head())
print("\nFemale Subset:\n", female_subset.head())
print("\nWhite Race Subset:\n", white_race_subset.head())
```

```python
# Merge two subsets

merged_subset = pd.concat([india_subset, female_subset])
print("\nMerged Subset (India + Female):\n", merged_subset.head())
```

```python
# Sort data by education-num

sorted_adult = adult.sort_values(by='education-num', ascending=False)
print("\nSorted by Education-Num:\n", sorted_adult.head())
```

```python
# Transpose first 5 rows

transposed = adult.head().T
print("\nTransposed Data (first 5 rows):\n", transposed)
```

```python
#  Melting Data to Long Format

melted = pd.melt(adult,
                 id_vars=['sex', 'race', 'native-country'],
                 value_vars=['age', 'education', 'hours-per-week'],
                 var_name='attribute',
                 value_name='value')

print("\nMelted Data:\n", melted.head())
```

```python
# Casting to Wide Format

casted = melted.pivot_table(index=['race', 'sex'],
                            columns='attribute',
                            values='value',
                            aggfunc=lambda x: x.iloc[0] if x.dtype == 'O' else x.mean())

print("\nCasted Data (Wide Format):\n", casted.head())
```
