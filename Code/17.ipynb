Visualize the Heart disease dataset by plotting the following graphs using Python. (Define 
objective for every graph)  
a. Histograms    
b. Dot Plots   
c. Bar Plots   
d. Line Charts   
e. Add Histogram and Scatter plot to box plot.  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('../files/heart.csv')

# # Check again

# Set seaborn style
sns.set(style="whitegrid")

# a. Histograms
# Objective: Show the distribution of the 'age' column to understand the age range of patients.
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Dot Plots
# Objective: Show the distribution of 'thalach' (maximum heart rate achieved) across individuals
plt.figure(figsize=(10, 6))
sns.stripplot(x='target', y='thalach', data=data, jitter=True, hue='target', palette='Set2', dodge=True, legend=False)
plt.title('Dot Plot of Maximum Heart Rate Achieved (thalach) by Heart Disease Presence')
plt.xlabel('Heart Disease Presence (0 = No, 1 = Yes)')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.show()
```

```python
# c. Bar Plots
# Objective: Show the count of individuals with and without heart disease ('target' column).
plt.figure(figsize=(10, 6))
sns.countplot(x='target', data=data, hue='target', palette='Set2', legend=False)
plt.title('Distribution of Heart Disease Presence')
plt.xlabel('Heart Disease Presence (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# d. Line Charts
# Objective: Show the relationship between age and maximum heart rate achieved (thalach) with a line plot.
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='thalach', data=data, color='purple')
plt.title('Age vs Maximum Heart Rate Achieved')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.show()
```

```python
# e. Add Histogram and Scatter plot to Box Plot
# Objective: Combine histogram, scatter plot, and box plot to visualize the relationship between age and heart disease presence.

# Set up the figure
plt.figure(figsize=(12, 6))

# Box Plot
plt.subplot(1, 2, 1)
sns.boxplot(x=data['target'], y=data['age'], color='lightblue')
plt.title('Box Plot of Age by Heart Disease Presence')
plt.xlabel('Heart Disease Presence')
plt.ylabel('Age')
```

```python

# Histogram
plt.subplot(1, 2, 2)
sns.histplot(data['age'], bins=20, kde=True, color='lightgreen')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot with Heart Disease Presence
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['age'], y=data['thalach'], hue=data['target'], palette='Set1')
plt.title('Age vs Maximum Heart Rate Achieved (thalach) by Heart Disease Presence')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate (thalach)')
plt.legend(title='Heart Disease Presence', labels=['No', 'Yes'])
plt.show()
```
