Visualize the Heart disease dataset by plotting the following graphs using Python. (Define 
objective for every graph)  
a. Histograms    
b. Pie Charts   
c. Box Plots   
d. Scatter Plots   
e. Add boxplots to a scatterplot  

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Assuming the dataset is named 'heart_disease.csv')
df = pd.read_csv('../files/heart.csv')

# a. Histograms
df.hist(figsize=(10, 8), bins=20)
plt.suptitle("Histograms of Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python

# b. Pie Charts
target_counts = df['target'].value_counts()  # Assuming 'target' is the heart disease column
target_counts.plot.pie(autopct='%1.1f%%', figsize=(8, 8), startangle=90)
plt.title("Pie Chart of Heart Disease Distribution")
plt.ylabel("")
plt.show()
```

```python
# c. Box Plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, width=0.5)
plt.title("Box Plots of Features")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# d. Scatter Plots
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['age'], y=df['chol'], hue=df['target'], palette='coolwarm', alpha=0.7)
plt.title("Scatter Plot: Age vs Cholesterol")
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

```python
# e. Add Box Plots to a Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['age'], y=df['chol'], hue=df['target'], palette='coolwarm', alpha=0.7)
sns.boxplot(x=df['target'], y=df['chol'], color='lightgray', width=0.5, fliersize=5, showfliers=False)
plt.title("Box Plot Overlayed on Scatter Plot: Age vs Cholesterol")
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```
