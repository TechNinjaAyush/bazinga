```python
import pandas as pd
```

```python
#load Iris dataset

iris = pd.read_csv('../files/iris.csv')
iris.head(10)
```

```python
# Create data subsets for different species

setosa = iris[iris['variety'] == 'Setosa']
versicolor = iris[iris['variety'] == 'Versicolor']
virginica = iris[iris['variety'] == 'Virginica']

print("\nSetosa Subset:\n")
setosa.head()
```

```python
print("\nVersicolor Subset:\n")
versicolor.head()
```

```python
print("\nVirginica Subset:\n")
virginica.head()
```

```python
# Merge two subsets (Setosa and Versicolor)

merged_subset = pd.concat([setosa, versicolor], axis=0)
merged_subset.head()
```

```python
# Sort Data by Petal Length

sorted_by_petal_length = iris.sort_values(by='petal.length')
sorted_by_petal_length.head()
```

```python
# Transposing Data

transposed_data = iris.T
transposed_data
```

```python
# Melting Data to Long Format

melted_data = pd.melt(iris, id_vars=['variety'], var_name='measurement', value_name='value')
melted_data.head()
```

```python
# Casting Data to Wide Format
casted_data = melted_data.pivot_table(index=['variety'], columns='measurement', values='value', aggfunc='mean').reset_index()
```

```python
print("\nCasted (Wide Format):\n", casted_data.head())
```
